```python
# Hierarchical Clustering

# Creating a Dendrogram
# SciPy linkage() function performs hierarchical clustering on an array of samples.

# Importing the libraries
# pip3 install pandas
# pip3 install scipy
# pip3 install matplotlib
import pandas as pd
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

# Load df
seeds_df = pd.read_csv('/Users/alexandergursky/Local_Repository/Datasets/Structured/CSV/Grains/seeds_appended.csv')
```

```python
# Creating an array for the clustering model
seeds_arr = seeds_df.iloc[:,0:8].values

# Creating a list for the seed names
species = seeds_df['names'].values.tolist()
```

```python
# Calculate the linkage: mergings
mergings = linkage(seeds_arr, method='complete')

# Plot and view the Dendrogram
# I had to adjust the size of the figure so I did the fig, ax =.... thing to do so. You don't actually need that to make a Dendrogram. Data is too large.
fig, ax = plt.subplots(figsize=(80,20))
dendrogram(mergings, labels=species, leaf_rotation=90, leaf_font_size=17, ax=ax)

plt.show()
```
