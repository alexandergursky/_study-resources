```python
# Locating directional data variance 

# Imports
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA


# Load df
grains_df = pd.read_csv('/Users/alexandergursky/Local_Repository/Datasets/Structured/CSV/Grains/seeds-width-vs-length.csv', header= None)

# creating an array for PCA
grains_arr = grains_df.values

# Make a scatter plot of the untransformed points
plt.scatter(grains_arr[:,0], grains_arr[:,1])
```

```python
# Make a second scatter plot of the untransformed points to view the direction that the data varies the most
plt.scatter(grains_arr[:,0], grains_arr[:,1])

# Create a PCA instance: model
model = PCA()

# Fit model to points
model.fit_transform(grains_arr)

# Get the mean of the grain samples: mean
mean = model.mean_

# Get the first principal component: first_pc
first_pc = model.components_[0,:]

# Plot first_pc as an arrow, starting at mean
plt.arrow(mean[0], mean[1], first_pc[0], first_pc[1], color='red', width=0.01)

# Keep axes on same scale
plt.axis('equal')
plt.show()

# This is the direction in which the grain data varies the most.
```
