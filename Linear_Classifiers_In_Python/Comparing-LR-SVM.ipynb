# Comparing Logistic Regression and Support Vector Machine

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn import datasets
from sklearn.model_selection import train_test_split

# Loading data
digits = datasets.load_digits()

# Making train test split
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target)
```

```python
# Comparison

# Apply logistic regression and print scores
lr = LogisticRegression()
lr.fit(X_train, y_train)
print(lr.score(X_train, y_train))  # Training accuracy
print(lr.score(X_test, y_test))  # Validation accuracy

# Apply SVM and print scores
svm = SVC()
svm.fit(X_train, y_train)
print(svm.score(X_train, y_train))  # Training accuracy
print(svm.score(X_test, y_test))  # Validation accuracy
```
