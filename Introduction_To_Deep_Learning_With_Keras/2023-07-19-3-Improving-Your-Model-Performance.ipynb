## Learning curves
  
Learning curves provide a lot of information about your model. Now that you know how to use the `History` callback to plot them, you will learn how to read them to get the most value out of them. So far we've seen two types of learning curves: loss curves and accuracy curves.
  
<img src='../_images/learning-curves-in-neural-networks.png' alt='img' width='740'>
  
**Loss curve**
  
Loss tends to decrease as epochs go by. This is expected since our model is essentially learning to minimize the loss function. Epochs are shown on the X axis and loss on the Y-axis. As epochs go by our loss value decreases. After a certain amount of epochs, the value converges, meaning it no longer gets much lower than that. We've arrived at a minimum.
  
<img src='../_images/learning-curves-in-neural-networks1.png' alt='img' width='740'>
  
**Accuracy curve**
  
Accuracy curves are similar but opposite in tendency if the Y-axis shows accuracy it now tends to increase as epochs go by. This shows that our model makes fewer mistakes as it learns.
  
<img src='../_images/learning-curves-in-neural-networks2.png' alt='img' width='740'>
  
**Overfitting**
  
If we plot training versus validation data we can identify overfitting. We will see the training and validation curves start to diverge. Overfitting is when our model starts learning particularities of our training data which don't generalize well on unseen data. The `tensorflow.keras.callbacks.EarlyStopping` `callback=` is useful to stop our model before it starts overfitting.
  
<img src='../_images/learning-curves-in-neural-networks3.png' alt='img' width='740'>
  
**Unstable curves**
  
But not all curves are smooth and pretty, many times we will find unstable curves. There are many reasons that can lead to unstable learning curves; the chosen `optimizer=`, `learning_rate=`, `batch_size=`, network architecture, weight initialization, etc. All these parameters can be tuned to improve our model learning curves, as we aim for better accuracy and generalization power. We will cover this in the following videos.
  
<img src='../_images/learning-curves-in-neural-networks4.png' alt='img' width='740'>
  
**Can we benefit from more data?**
  
Neural networks are well known for surpassing traditional machine learning techniques as we increase the size of our datasets. We can check whether collecting more data would increase a modelâ€™s generalization and accuracy. We aim at producing a graph like this one, where we have fitted our model with increasing amounts of training data and plotted the values for the training and test accuracies of each run. 
  
<img src='../_images/learning-curves-in-neural-networks5.png' alt='img' width='740'>
  
If after using all our data we see that our test still has a tendency to improve, that is, it's not parallel to our training set curve and it's increasing, then it's worth it to gather more data if possible to allow the model to keep learning.
  
<img src='../_images/learning-curves-in-neural-networks6.png' alt='img' width='740'>
  
**Coding train size comparison**
  
How would we go about coding a graph like the previous one? Imagine we want to evaluate an already built and compiled model and that we have partitioned our data into `X_train`, `y_train`, `X_test` and `y_test`. We first store the model initial weights, this is done by calling `.get_weights()` on our model, we then initialize two lists to store train and test accuracies.
  
<img src='../_images/learning-curves-in-neural-networks7.png' alt='img' width='740'>
  
**Coding train size comparison II**
  
We loop over a predefined list of train sizes and for each training size we get the corresponding training data fraction. Before any training, we make sure our model starts with the same set of weights by setting them to the `initial_weights=` using the `set_weights=` function. After that, we can fit our model on the training fraction. We use an EarlyStopping callback which monitors loss, but it's important to note that it's not validation loss since we haven't provided the `.fit()` method with validation data. After the training is done, we can get the accuracy for the training set fraction and the accuracy from the test set and append it to our lists of accuracies. Observe that the same quantity of test data observations were used to evaluate each iteration.

### Learning the digits
  
You're going to build a model on the **digits dataset**, a sample dataset that comes pre-loaded with scikit-learn. The digits dataset consist of **8x8 pixel handwritten digits from 0 to 9**:
  
<center><img src='../_images/learning-curves-in-neural-networks9.png' alt='img' width='740'></center>
  
You want to distinguish between each of the 10 possible digits given an image, so we are dealing with **multi-class classification**.
  
The dataset has already been partitioned into `X_train`, `y_train`, `X_test`, and `y_test`, using 30% of the data as testing data. The labels are already one-hot encoded vectors, so you don't need to use `tensorflow.keras.utils.to_categorical()` function.

Let's build this new model!
  
1. Add a `tensorflow.keras.layers.Dense()` layer of 16 neurons with `relu` `activation=` and an `input_shape=` that takes the total number of pixels of the 8x8 digit image.
2. Add a `tensorflow.keras.layers.Dense()` layer with 10 outputs and `softmax` `activation=`.
3. Compile your model with `'adam'`, `'categorical_crossentropy'`, and `'accuracy'` `metrics=`.
4. Make sure your model works by predicting on `X_train`.

````python
from keras.utils import to_categorical
from sklearn.model_selection import train_test_split


# Loading the numpy arrays required
X = np.load('../_datasets/digits_pixels.npy')
y = np.load('../_datasets/digits_target.npy')
print('Feature shape: {}\nTarget shape: {}'.format(X.shape, y.shape))

# Conversion of values into OHE format
y = to_categorical(y)
print('Categorical Target shape: {}'.format(y.shape))

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
````

<span style='color:#7393B3'>NOTE:</span>  
 
File format `.npy` is a binary file format that storys a single `numpy.array()` or object array

````python
from keras.models import Sequential
from keras.layers import Dense


# Initialize a Sequential model
model = Sequential()

# Input and hidden layer with input_shape, 16 neurons, and relu
model.add(Dense(16, activation='relu', input_shape=(64, )))

# Output layer with 10 neurons (one per digit) and softmax
model.add(Dense(10, activation='softmax'))

# Compile your model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Test if your model is well assembled by predicting before training
print(model.predict(X_train))
````

Predicting on training data inputs before training can help you quickly check that your model works as expected.

### Is the model overfitting?
  
Let's train the `model` you just built and plot its learning curve to check out if it's overfitting! You can make use of the loaded function `plot_loss()` to plot training loss against validation loss, you can get both from the `History` callback.
  
If you want to inspect the `plot_loss()` function code, paste this in the console: `show_code(plot_loss)`
  
1. Train your model for 60 epochs, using `X_test` and `y_test` as validation data.
2. Use `plot_loss()` passing `loss` and `val_loss` as extracted from the `.history` attribute of the `h_callback` object.

````python
# Function to plot the loss
def plot_loss(loss, val_loss):
    plt.figure()
    plt.plot(loss)
    plt.plot(val_loss)
    plt.title('Model loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Test'], loc='upper right')
````

````python
# Train your model for 60 epochs, using X_test and y_test as validation data
h_callback = model.fit(X_train, y_train, epochs=60, validation_data=(X_test, y_test), verbose=0)

# Extract from the h_callback object loss and val_loss to plot the learning curve
plot_loss(h_callback.history['loss'], h_callback.history['val_loss'])
````

### Do we need more data?
  
It's time to check whether the digits dataset model you built benefits from more training examples!
  
In order to keep code to a minimum, various things are already initialized and ready to use:
  
- The `model` you just built.
- `X_train`, `y_train`, `X_test`, and `y_test`.
- The `initial_weights` of your model, saved after using `model.get_weights()`.
- A pre-defined list of training sizes: `training_sizes`.
- A pre-defined early stopping callback monitoring loss: `early_stop`.
- Two empty lists to store the evaluation results: `train_accs` and `test_accs`.
  
Train your model on the different training sizes and evaluate the results on `X_test`. End by plotting the results with `plot_results()`.
  
1. Get a fraction of the training data determined by the `size` we are currently evaluating in the loop.
2. Set the model weights to the `initial_weights` with `.set_weights()` and train your model on the fraction of training data using `early_stop` as a callback.
3. Evaluate and store the accuracy for the training fraction and the test set.
4. Call `plot_results()` passing in the training and test accuracies for each training size.

````python
# Function to plot the comparison of accuracy metric to training size
def plot_results(train_accs, test_accs):
    plt.plot(train_accs, 'o-', label="Training Accuracy")
    plt.plot(test_accs, 'o-', label="Test Accuracy")
    plt.title('Accuracy vs Number of training samples')
    plt.xlabel('# of training samples')
    plt.ylabel('Accuracy')
    plt.legend(loc="best")
````

````python
from keras.callbacks import EarlyStopping


# Gathering the trained weights from the prior model to recursively reset (benchmark)
initial_weights = model.get_weights()

# Instantiation of callback
early_stop = EarlyStopping(monitor='loss', patience=1)

# Creating the loop variables 
train_accs, test_accs = [], []

# Creating the training sizes to compare if model benefits from more data
# NOTE [10%, 40%, 70%, 100%] of observations in train split. Math came out to [125.7, 502.9, 879.9, 1257] so I rounded up.
training_sizes = [126, 503, 880, 1257]

for size in training_sizes:
    # Get a fraction of training data (we only care about the training data)
    X_train_frac, y_train_frac = X_train[:size], y_train[:size]
    
    # Reset the model to the initial weights and train it on the new training data fraction
    model.set_weights(initial_weights)
    model.fit(X_train_frac, y_train_frac, epochs=50, callbacks=[early_stop])
    
    # Evaluate and store both: the training data fraction and the complete test set results
    train_accs.append(model.evaluate(X_train_frac, y_train_frac)[1])
    test_accs.append(model.evaluate(X_test, y_test)[1])


# Plot train vs test accuracies
plot_results(train_accs, test_accs)
````

<span style='color:#7393B3'>NOTE:</span>  If the test set accuracy is starting to flatten. Then your model would not benefit a lot from more training data.

## Activation functions
  
So far we've been using several activation functions in our models, but we haven't yet covered their role in neural networks other than when it comes to obtaining the output we want in our output layer.
  
**An activation function**
  
Inside the neurons of any neural network the same process takes place:
  
<img src='../_images/activation-functions-keras-neural-networks.png' alt='img' width='740'>
  
A summation of the inputs reaching the neuron multiplied by the weights of each connection and the addition of the bias weight. This operation results in a number: $a$, which can be anything, it is not bounded.
  
<img src='../_images/activation-functions-keras-neural-networks1.png' alt='img' width='740'>
  
We pass this number into an activation function that essentially takes it as an input and decides how the neuron fires and which output it produces. Activation functions impact learning time, making our model converge faster or slower and achieving lower or higher accuracy. They also allow us to learn more complex functions.
  
<img src='../_images/activation-functions-keras-neural-networks2.png' alt='img' width='740'>
  
**Activation zoo**

def sigmoid(x):

````python
from sklearn.model_selection import train_test_split


# X/y split
sensors = irrigation.drop(['parcel_0', 'parcel_1', 'parcel_2'], axis=1).to_numpy()
parcels = irrigation[['parcel_0', 'parcel_1', 'parcel_2']].to_numpy()

# X/y shapes
print('X.shape: {}\ny.shape: {}\n'.format(sensors.shape, parcels.shape))

# Input/output shapes required for the ANN
print('Input features: {}\nOutput targets: {}'.format(sensors.shape[1], parcels.shape[1]))

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(sensors, parcels, test_size=0.3, stratify=parcels)
````

````python
# SEED
np.random.seed(1)

# Function to return a new model with the given activation
def get_model(act_function=str):
    '''This function is used to recursively select a `tensorflow.nn` or `tensorflow.keras.activations` activation function.
    `tensorflow.keras.layers.Dense(activation='')` does not contain all available activation functions by default.'''

    # Model instantiation
    model = Sequential()
    # Conditions, NOTE that you can just pass a string to param vs using tensorflow.nn or keras.activations
    if act_function == 'relu':              # Rectified Linear Unit (ReLU)
        model.add(Dense(64, activation=tf.nn.relu, input_shape=(20, )))
    elif act_function == 'leaky_relu':      # Leaky Rectified Linear Unit (Leaky ReLU)
        model.add(Dense(64, activation=tf.nn.leaky_relu, input_shape=(20, )))
    elif act_function == 'sigmoid':         # Sigmoid activation function (Sigmoid)
        model.add(Dense(64, activation=tf.nn.sigmoid, input_shape=(20, )))
    elif act_function == 'tanh':            # Hyperbolic Tangent (Tanh)
        model.add(Dense(64, activation=tf.nn.tanh, input_shape=(20, )))
    elif act_function == 'elu':             # Exponential Linear Unit (ELU)
        model.add(Dense(64, activation=tf.nn.elu, input_shape=(20, )))
    elif act_function == 'gelu':            # Gaussian error linear unit (GELU)
        model.add(Dense(64, activation=tf.nn.gelu, input_shape=(20, )))
    elif act_function == 'relu6':           # Rectified Linear Unit 6 (ReLU6)
        model.add(Dense(64, activation=tf.nn.relu6, input_shape=(20, )))
    elif act_function == 'selu':            # Scaled Exponential Linear Unit (SELU)
        model.add(Dense(64, activation=tf.nn.selu, input_shape=(20, )))
    elif act_function == 'silu':            # Sigmoid-Weighted Linear Unit (SiLU)
        model.add(Dense(64, activation=tf.nn.silu, input_shape=(20, )))
    elif act_function == 'swish':           # Swish activation function (Swish)
        model.add(Dense(64, activation=tf.nn.swish, input_shape=(20, )))
    elif act_function == 'linear':          # Linear activation function (pass-through)
        model.add(Dense(64, activation=keras.activations.linear, input_shape=(20, )))
    elif act_function == 'mish':            # Mish activation function (Mish)
        model.add(Dense(64, activation=keras.activations.mish, input_shape=(20, )))
    elif act_function == 'softplus':        # Softplus activation function (Softplus)
        model.add(Dense(64, activation=keras.activations.softplus, input_shape=(20, )))
    elif act_function == 'softsign':        # Softsign activation function (Softsign)
        model.add(Dense(64, activation=keras.activations.softsign, input_shape=(20, )))
    elif act_function == 'softmax':         # Softmax activation function (Softmax)
        model.add(Dense(64, activation=keras.activations.softmax, input_shape=(20, )))
    elif act_function == 'exponential':     # Exponential activation function (Softmax)
        model.add(Dense(64, activation=keras.activations.exponential, input_shape=(20, )))
    elif act_function == 'hard_sigmoid':    # Hard Sigmoid activation function (Hard Sigmoid)
        model.add(Dense(64, activation=keras.activations.hard_sigmoid, input_shape=(20, )))
    # NOTE That there are more than these, but these are the most commonly seen throughout deep learning
    # Add more activation functions as needed...
    
    # Errors to raise if not found, prevention of RecursionError
    else:
        if type(act_function) != str:
            raise TypeError('`get_model()` must accept a valid activation function as a datatype str')
        else:
            raise ValueError('Activation function passed is not present in `get_model()`.')

    # Add an output layer of 3 neurons with sigmoid activation
    model.add(Dense(3, activation='sigmoid'))
    # Compile your model with binary crossentropy loss
    model.compile(optimizer='adam',loss='binary_crossentropy', metrics=['accuracy'])

    return model
    
````

````python
import time


# Activation functions to try
activations = [
    'relu', 'leaky_relu', 'sigmoid', 'tanh', 'elu',
    'gelu', 'relu6', 'selu', 'silu', 'swish',
    'linear', 'mish', 'softplus', 'softsign', 'softmax', 'exponential', 'hard_sigmoid'
]

# Loop over the activation functions
activation_results = {}

for act in activations:
    # Stopwatch
    start_time = time.time()
    print('Started training for: {}...'.format(act))
    
    # Get a new model with the current activation
    model = get_model(act)

    # Fit the model and store the History results
    h_callback = model.fit(X_train, y_train, epochs=100, validation_data=(X_test, y_test), verbose=0)
    activation_results[act] = h_callback

    # Stopwatch ended
    end_time = time.time()
    elipsed_time = end_time - start_time

    print('Finished training with: {}, Training Time: {:.2f} seconds'.format(act, elipsed_time))
````

## Batch size and batch normalization
  
Itâ€™s time to learn the concepts of batch size and batch normalization.
  
**Batches**
  
A mini-batch is a subset of data samples. If we were training a neural network with images, each image in our training set would be a sample and we could take mini-batches of different sizes from the training set batch.
  
**Mini-batch**
  
Remember that during an epoch we feed our network, calculate the errors and update the network weights. It's not very practical to update our network weights only once per epoch after looking at the error produced by all training samples. In practice, we take a mini-batch of training samples. And that way, if our training set has 9 images and we choose a `batch_size=` of 3, we will perform 3 weight updates per epoch, one per mini-batch.
  
**Mini-batches**
  
Networks tend to train faster with mini-batches since weights are updated often. Sometimes datasets are so huge that they would struggle to fit in RAM memory if we didn't use mini-batches. Also, the noise produced by a small batch-size can help escape local minima. A couple of disadvantages are the need for more iterations and finding a good batch size.
  
Advantages:  
- Networks train faster (more weight updates in same amount of time)
- Less RAM memory required, can train on huge datasets
- Noise can help networks reach a lower error, escaping local minima
  
Disadvantage:  
- More iterations need to be run
- Need to be adjusted, we need to find a good batch size
  
**Effects of batch sizes**
  
Here you can see how different batch sizes converge towards a minimum as training goes by. Training with all samples is shown in blue. Mini-batching is shown in green. Stochastic gradient descent, in red, uses a `batch_size=` of 1. We can see how the path towards the best value for our weights is noisier the smaller the `batch_size=`. They reach the same value after a different number of iterations.
  
<img src='../_images/effects-of-batch-size-tuning-neural-networks.png' alt='img' width='740'>
  
**Batch size in Keras**
  
You can set your own `batch_size=` with the `batch_size=` parameter on the model's `.fit()` method. 
  
- Keras uses a default batch-size of 32.  
- Increasing powers of two tend to be used.  
- As a rule of thumb, you tend to make your batch size bigger the bigger your dataset.
  
**Normalization in machine learning**
  
Normalization (<span style='color:#7393B3'>NOTE:</span> *the formula they are discussing here is actually standardization*) is a common pre-processing step in machine learning algorithms, especially when features have different scales. One way to normalize data is to subtract its mean value and divide by the standard deviation. We always tend to normalize our model inputs. This avoids problems with activation functions and gradients.
  
<img src='../_images/effects-of-batch-size-tuning-neural-networks1.png' alt='img' width='740'>
  
This leaves everything centered around 0 with a standard deviation of 1.
  
<img src='../_images/effects-of-batch-size-tuning-neural-networks2.png' alt='img' width='740'>
  
**Reasons for batch normalization**
  
Normalizing neural networks inputs improve our model. But deeper layers are trained based on previous layer outputs and since weights get updated via gradient descent, consecutive layers no longer benefit from normalization and they need to adapt to previous layers' weight changes, finding more trouble to learn their own weights. Batch normalization makes sure that, independently of the changes, the inputs to the next layers are normalized. It does this in a smart way, with trainable parameters that also learn how much of this normalization is kept scaling or shifting it.
  
<img src='../_images/effects-of-batch-size-tuning-neural-networks3.png' alt='img' width='740'>
  
**Batch normalization advantages**
  
This improves gradient flow, allows for higher learning rates, reduces weight initializations dependence, adds regularization to our network and limits internal covariate shift; which is a funny name for a layer's dependence on the previous layer outputs when learning its weights. Batch normalization is widely used today in many deep learning models.
  
Advantages:  
- Improves gradient flow
- Allows higher learning rates
- Reduces dependence on weight initializations
- Acts as an unintended form of regularization
- Limits inter covariate shift
  
**Batch normalization in Keras**
  
Batch normalization in Keras is applied as a layer. So we can place it in between two layers. We import batch normalization from `tensorflow.keras.layers.BatchNormalization`. We then instantiate a `Sequential()` model, add an input layer, and then add a batch normalization layer. We finalize this binary classification model with an output layer.
  


### Changing batch sizes
  
You've seen models are usually trained in batches of a fixed size. The smaller a batch size, the more weight updates per epoch, but at a cost of a more unstable gradient descent. Specially if the batch size is too small and it's not representative of the entire training set.
  
Let's see how different batch sizes affect the accuracy of a simple binary classification model that separates red from blue dots.
  
You'll use a batch size of one, updating the weights once per sample in your training set for each epoch. Then you will use the entire dataset, updating the weights only once per epoch.
  
1. Use `get_model()` to get a new, already compiled, model, then train your model for 5 epochs with a `batch_size=` of 1.
2. Now train a new model with `batch_size=` equal to the size of the training set.

````python
dots = pd.read_csv('../_datasets/dots.csv')
print(dots.shape)
dots.head()
````

````python
# X/y split
X = dots.iloc[:, :-1]
y = dots.iloc[:, -1]
print('X.shape: {}\ny.shape: {}'.format(X.shape, y.shape))

# Train/test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
````

````python
# Defining a function to reset/benchmark the model
def get_model():
    model = Sequential()
    model.add(Dense(4, activation='relu', input_shape=(2, )))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(optimizer='sgd', loss='binary_crossentropy', metrics=['accuracy'])

    return model
````

````python
# Get a fresh new model with get_model
model = get_model()

# Train your model for 5 epochs with a batch size of 1
model.fit(X_train, y_train, epochs=5, batch_size=1)
print("\nThe accuracy when using a batch size of 1 is {:.2f}".format(model.evaluate(X_test, y_test)[1]))
````

````python
# Getting a new model
model = get_model()

# Fit your model for 5 epochs with a batch of size the training set
model.fit(X_train, y_train, epochs=5, batch_size=X_train.shape[0])
print("\n The accuracy when using the whole training set as batch-size was: {:.2f}".format(
    model.evaluate(X_test, y_test)[1]))
````

You can see that accuracy is lower when using a batch size equal to the training set size. This is not because the network had more trouble learning the optimization function: Even though the same number of epochs were used for both batch sizes the number of resulting weight updates was very different! With a batch the size of the training set and 5 epochs we only get 5 updates total, each update computes and averaged gradient descent with all the training set observations. To obtain similar results with this batch size we should increase the number of epochs so that more weight updates take place.

### Preparing a model for tuning
  
Let's tune the hyperparameters of a binary classification model that does well classifying the **breast cancer dataset**.
  
You've seen that the first step to turn a model into a `sklearn` estimator is to build a function that holds/creates it. The definition of this function is important since hyperparameter tuning is carried out by varying the arguments your function receives.
  
Build a simple `create_model()` function that receives both a `learning_rate=` and an `activation=` function as arguments. The Adam optimizer has been imported as an object from `tensorflow.keras.optimizers.Adam` so that you can also change its `learning_rate=` parameter.
  
1. Set the `learning_rate=` of the `Adam` optimizer object to the one passed in the arguments.
2. Set the hidden layers activations to the one passed in the arguments.
3. Pass the `optimizer=` and the `'binary_crossentropy'` `loss=` to the `.compile()` method.

````python
# Creates a model given an activation and learning rate
def create_model(learning_rate='learning_rate', activation='model__activation'):
    # Create an Adam optimizer with the given learning rate
    opt = tf.keras.optimizers.Adam(learning_rate=learning_rate)
    
    # Create your binary classification model
    model = Sequential()
    model.add(Dense(128, input_shape=(30, ), activation=activation))
    model.add(Dense(256, activation=activation))
    model.add(Dense(1, activation='sigmoid'))
    
    # Compile your model with your optimizer, loss and metrics
    model.compile(optimizer=opt, loss='binary_crossentropy', metrics=['accuracy'])
    
    return model

````

With this function ready you can now create a `sklearn` estimator and generate different models to perform simple hyperparameter tuning on!

````python
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import KFold


# Create a KerasClassifier
model = KerasClassifier(model=create_model, loss='binary_crossentropy', learning_rate='learning_rate', verbose=0)

# Define the parameters to try out
params = {
    'model__activation': ['relu', 'tanh'],
    'batch_size': [32, 128, 256],
    'epochs': [50, 100, 200],
    'learning_rate': [0.1, 0.01, 0.001]
}

# Create a RandomizedSearchCV object passing in the parameters to try out
random_search = RandomizedSearchCV(
    estimator=model,                # Model
    param_distributions=params,     # Parameters to test
    cv=KFold(n_splits=3),           # KFold Cross Validation, ie. Validation data
    n_iter=10,                      # Number of models to attempt
    n_jobs=-1,                      # Multiprocessing with cores
    verbose=0                       # Verbosity
)

# Training
random_search_results = random_search.fit(X, y)

# Print results
print("Best score: {:.4f} using {}\n".format(random_search_results.best_score_, random_search_results.best_params_))

# Print parameters tested
means = random_search_results.cv_results_['mean_test_score']
stds = random_search_results.cv_results_['std_test_score']
params = random_search_results.cv_results_['params']

for mean, stdev, param in zip(means, stds, params):
    print("{:.4f} ({:.4f}) with: {}".format(mean, stdev, param))
````

<span style='color:#7393B3'>NOTE:</span>  
  
Grid search and random search are hyperparameter optimization techniques used to find the best set of hyperparameters for a machine learning model. On the other hand, TensorFlow callbacks are mechanisms to monitor and customize the behavior of a model during training. While both approaches are used in model optimization, they serve different purposes.
  
1. **Grid Search and Random Search:**  
  
**Grid Search:** Grid search is an exhaustive search technique where you define a grid of hyperparameter values to explore. It systematically tries all possible combinations of hyperparameters from the grid and evaluates the model performance using cross-validation. Grid search is computationally expensive as it explores all combinations, but it guarantees to find the best hyperparameters from the search space.
  
**Random Search:** Random search is a more efficient alternative to grid search. Instead of exploring all possible combinations, it randomly samples hyperparameters from predefined distributions. It performs a limited number of iterations, making it more scalable and less computationally expensive than grid search. While random search may not guarantee the best hyperparameters, it often discovers good configurations faster.
  
2. **TensorFlow Callbacks:**  
  
TensorFlow callbacks are functions that are executed at specific points during the model training process. They allow you to monitor the training progress, take actions based on certain conditions, and modify the model behavior. Some common TensorFlow callbacks include:
  
- `ModelCheckpoint`: Saves the model's weights during training, allowing you to restore the best model later.
- `EarlyStopping`: Stops training if a monitored metric stops improving, preventing overfitting and saving time.
- `ReduceLROnPlateau`: Adjusts the learning rate if the model's performance plateaus, fine-tuning the optimization process.
- `TensorBoard`: Logs training information for visualization in TensorFlow's TensorBoard.
  
**Combining Grid Search/Random Search with TensorFlow Callbacks:**
  
In practice, you can use grid search or random search to find the best hyperparameters, and then utilize TensorFlow callbacks to monitor and optimize the training process based on the chosen hyperparameters. For example, you can use grid search to find the best learning rate, batch size, and activation function, and then use early stopping and model checkpoint callbacks during the training process to improve the model's performance and save the best model.
  
It's essential to note that hyperparameter optimization is often an iterative process. You can start with a coarse grid or random search to narrow down the search space and then perform a finer-grained search around promising regions to further improve the model's performance.
  
Overall, grid search and random search are used for hyperparameter optimization, while TensorFlow callbacks are used for monitoring and customizing the model training process. Combining both approaches can lead to more efficient and effective model optimization.
